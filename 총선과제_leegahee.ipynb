# 6. 하노이 탑 옮기기

원반이 n개인 하노이탑을 옮기기 위한 원반 이동순서 출력하는 알고리즘 만들기  
하노이 탑을 알아보고 재귀호출을 이용해 문제를 풀어보자  

알고리즘 Hanoi(개수,시작점,끝나는점,경유지)  
* 종료조건: 원반이 한개일때
***
* n개일때
1. (1) -- n-1 --> (2)
2. (1) -- 1------------------>(3)
3. ......................(2)-- n-1--->(3)

```python
# 생각해보기 
def hanoi(n,start,end,via):
    if n == 1 :
        print(start,via,end) 
    else :
        hanoi(n-1,start,via,end)
        print(start,via,end) 
        print('나머지것을 다 중간으로 옮김')
        hanoi(1,start,end,via)
        # print(start,via,end) 
        hanoi(n-1,via,end,start)
        print(start,via,end) 
```

```python
hanoi(1,'a','b','c')
```
