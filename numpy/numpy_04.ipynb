```python
# 지수함수

x = np.arange(0,10)
y1 = x
y2 = x**2

plt.plot(x,y1,label='1차 함수')
plt.plot(x,y2,label='2차 함수')
plt.legend(loc='upper left')
plt.title('지수함수')
plt.show()
```

```python
# 시그모이드 함수 (0,1)

def sigmoid(x):
    s = 1 /(1+np.exp(-x))
    ds = s*(1-s) #미분공식
    return s,ds
```

```python
x1 = np.linspace(-10,10,100)
y1,y2 = sigmoid(x)
plt.plot(x1,y1)
plt.plot(x1,y2)
plt.show()
```
